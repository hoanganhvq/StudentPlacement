```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
df = pd.read_csv("data/placement_data.csv")
# 1. Tạo bản sao dữ liệu và mã hóa nhãn nếu chưa có
df_cls_analysis = df.copy()
if 'placement_status_num' not in df_cls_analysis.columns:
    df_cls_analysis['placement_status_num'] = df_cls_analysis['placement_status'].map({'Placed': 1, 'Not Placed': 0})

# 2. Tính tương quan và lọc ra cột target
corr_cls = df_cls_analysis.select_dtypes(include=[np.number]).corr()
cls_target_corr = corr_cls[['placement_status_num']].sort_values(by='placement_status_num', ascending=False)

# 3. Vẽ Heatmap
sns.heatmap(cls_target_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
plt.title("Tương quan của các Feature với khả năng TRÚNG TUYỂN (Classification)")
plt.show()

print("Top 5 yếu tố quyết định việc trúng tuyển:")
print(cls_target_corr.head(6)) # Bao gồm cả chính nó
```
